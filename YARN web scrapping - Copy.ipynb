```python
import requests as rq
url ="https://www.yarnplaza.com/yarn?br=23%2c34%2c38&page=2&limit=filterbr,filterft3,filterft10,filterft11,filterft37"
res = rq.get(url)
```

```python
from bs4 import BeautifulSoup as bs #import the libraries
yarns = bs(res.content)  #get content from page
print(yarns)
```

```python
results = yarns.find(id ="upProducts")
print(results)
```

```python
c_results = results.find_all("div", class_= "productlist-mainholder")
c_results
```

```python
c_resuts1= results.find_all("h3",class_="productlist-title gtm-product-impression")
yarn_titles = [title.text.strip() for title in c_resuts1] #get all yarn 

yarn_titles
```
