```python
import requests as rq
url ="https://yarnplaza.com/"
res = rq.get(url)
```

```python
from bs4 import BeautifulSoup as bs #import the libraries
yarns = bs(res.content)  #get content from page
print(yarns)
```

```python
results = yarns.find(id ="websiteblok29")
print(results)
```

```python
c_results = results.find_all("div", class_= "innerproductlist")
c_results
```

```python
c_resuts1= results.find_all("h3",class_="productlist-title gtm-product-impression")
yarn_titles = [t.text.strip() for t in c_resuts1] #get all yarn titles

yarn_titles
```
